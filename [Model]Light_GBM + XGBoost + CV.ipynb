# < Boosting, XGBoost, Light GBM >
1. 부스팅이란?
    - 배깅이 여러 개의 단일 모델을 만든 뒤 붓스트랩을 통해 데이터를 추출하여 모델 훈련.
    - 부스팅은 붓스트랩을 통해 데이터를 추출하고 다수의 분류기를 생성한다는 점에서 배깅과 유사
    - 앙상블 부스팅은 배깅과 다르게 앞에서 예측한 분류기가 틀린 부분에 있어 가중치를 부여
    - 즉, 배깅의 변형으로 모델이 잘 예측 못하는 부분을 개선
--------------------------------------------------------------------------------
    
2. 부스팅 알고리즘 종류
    1. AdaBoost : 다수결을 통한 정답 분류 및 오답에 가중치 부여
    2. GBM : Loss Function의 gradient를 통해 오답에 가중치 부여(반복 수행을 통해 오류를 최소화 할 수 있도록 가중치의 업데이트 값을 도출하는 기법)

    3. Xgboost : 병렬 처리 사용, 유연성이 좋다.(다양한 커스텀 최적화 옵션 제공), GBM에 기반하지만, GBM의 단점인 느린 수행 시간 및 과적합 규제 부재 등의 문제를 해결하여
각광 받고 있는 알고리즘

    4. Light GBM : 리프 중심 트리 분할로 더 빠른 학습 & 예측 수행 시간과 적은 메모리 사용량의 장점을 지님
    
--------------------------------------------------------------------------------
3. XGBoost란?
    - 트리 기반의 앙상블 학습에서 가장 각광 받고 있는 알고리즘 중 하나
    - 분류에 있어서 일반적으로 다른 머신러닝보다 뛰어난 예측 성능 나타냄
    - GBM에 기반
    - GBM의 단점인 느린 수행 시간 / 과적합 규제 부재 등의 문제 해결
    - 병렬 CPU 환경에서 병렬 학습이 가능해 기존 GBM보다 빠르게 학습 완료

--------------------------------------------------------------------------------
4. Light GBM이란?
    - 기존 트리 기반 알고리즘에서 사용하는 균형 트리 분할 방식을 사용하지 않고 리프 중심 트리 분할 방식을 사용
    - Light GBM은 적은 데이터에 대한 과적합이 발생하기 쉽다. 따라서 10,000건 이하의 데이터에 대해서는 조심해야 함
    - 과적합을 방지하고 성능을 보다 객관적으로 보기 위해 교차검증 사용

```python
data_dir = Path('/data')
feature_dir = Path('/data')
sub_dir = Path('/data')

trn_file = data_dir / 'train.csv'
tst_file = data_dir / 'test.csv'
sample_file = data_dir / 'sample_submission.csv'

target_col = 'class'
seed = 42
```

```python
algo_name = 'lgb'
feature_name = 'feature'
model_name = f'{algo_name}_{feature_name}'

feature_file = feature_dir / f'{feature_name}.csv'
sub_file = sub_dir / f'{model_name}.csv'
```

```python
df = pd.read_csv(feature_file, index_col=0)
print(df.shape)
df.head()
```

```python
y = df[target_col].values[:320000]
df.drop(target_col, axis=1, inplace=True)
trn = df.iloc[:320000]
tst = df.iloc[320000:]
feature_name = df.columns.tolist()
print(y.shape, trn.shape, tst.shape)
```

## <학습/검증 데이터 구분>

```python
X_trn, X_val, y_trn, y_val = train_test_split(trn, y, test_size=.2, random_state=seed)
print(X_trn.shape, X_val.shape, y_trn.shape, y_val.shape)
```

## <LightGBM 모델 학습>

```python
clf = lgb.LGBMClassifier(objective='multiclass',
                         n_estimators=1000,
                         num_leaves=64,
                         learning_rate=0.1,
                         min_child_samples=10,
                         subsample=.5,
                         subsample_freq=1,
                         colsample_bytree=.8,
                         random_state=seed,
                         n_jobs=-1)
clf.fit(X_trn, y_trn,
        eval_set=[(X_val, y_val)],
        eval_metric='multiclass',
        early_stopping_rounds=10)
p_val = clf.predict(X_val)
p_tst = clf.predict(tst)
```

```python
print(f'{accuracy_score(y_val, p_val) * 100:.4f}%')
```

## <피처 중요도 시각화>

```python
imp = pd.DataFrame({'feature': trn.columns, 'importance': clf.feature_importances_})
imp = imp.sort_values('importance').set_index('feature')
imp.plot(kind='barh')
```

- 피처셋을 바꿔보진 않았으나 현재의 피처셋으로는 Light GBM의 성능이 가장 좋음
- Random Forest와 Light GBM 모두 n_estimators를 크게 늘려보진 않음
- light GBM의 성능이 가장 좋으므로 우선 light GBM에 맞는 피처셋을 찾아볼 것.
